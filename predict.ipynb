```python
from ultralytics import YOLO
import cv2
import random

# YOLOv8 modelini yükle
# model = YOLO(r"C:\\Users\\denem\\Documents\\python\\receipt\\runs-20250301T094915Z-001\\runs\\detect\\train\\weights\\best.pt") 
model=YOLO(r"models\runs_v3-20250305T084703Z-001\runs_v3\detect\train3\weights\best.pt") # yeni

# 5 sabit renk belirleyelim (kutular için)
class_colors = {
    0: (0, 255, 0),     # Yeşil
    1: (0, 0, 255),     # Kırmızı
    2: (255, 0, 0),     # Mavi
    3: (0, 255, 255),   # Sarı
    4: (255, 255, 0)    # Turuncu
}

def predict_image(image_path,i):
    # Görüntüyü yükle
    image = cv2.imread(image_path)

    # YOLOv8 ile tahmin yap
    best_boxes = {}
    results = model(image)
    print(f"----------------{i}--{image_path}-------------------")
    print("\n--- Tahmin Sonuçları ---")

    for result in results:
        for box in result.boxes:
            class_id = int(box.cls[0])  # Sınıf ID
            conf = box.conf[0].item()   # Güven skoru
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Koordinatlar

            # Eğer bu sınıf ilk kez tespit ediliyorsa veya daha yüksek güvene sahipse güncelle
            if class_id not in best_boxes or conf > best_boxes[class_id][0]:
                best_boxes[class_id] = (conf, (x1, y1, x2, y2))

    # En iyi kutuları çiz
    for class_id, (conf, (x1, y1, x2, y2)) in best_boxes.items():
        label = f"{model.names[class_id]}: {conf:.2%}"  # Yüzdelik format (%)
        print(label)  # Konsola yazdır

        # Kutular için sabit renk belirle
        color = class_colors.get(class_id, (0, 255, 0))

        # Kutuyu çiz
        cv2.rectangle(image, (x1, y1), (x2, y2), color, 5)

        # Yazı için kontrastı artırmak amacıyla arka plan ekle
        label_size = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 1, 2)[0]
        label_x = x1
        label_y = y1 - 10
        # cv2.rectangle(image, (label_x, label_y - label_size[1]), (label_x + label_size[0], label_y + 5), color, -1)  # Arka plan
        # cv2.putText(image, label, (label_x, label_y), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)  # Beyaz yazı

    # Görüntüyü ekrana sığdır
    max_size = 800  # Maksimum genişlik/yükseklik
    height, width = image.shape[:2]

    if max(height, width) > max_size:
        scale = max_size / max(height, width)
        new_size = (int(width * scale), int(height * scale))
        image = cv2.resize(image, new_size, interpolation=cv2.INTER_AREA)

    # Görüntüyü göster
    cv2.imshow(f"{i}--{image_path}", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

for i in range(
  #  len(files)
 10
    ):
    predict_image(f"{path}{files[i]}",i)  


# predict_image(f"{path}{files[0]}",0)   yolo mobil 
```
