```python
import cv2,os
import json
from google.cloud import vision
from google import genai
```

```python


def detect_text_from_image(image_path):
    client = vision.ImageAnnotatorClient()
    
    with open(image_path, "rb") as image_file:
        content = image_file.read()
    
    image = vision.Image(content=content)
    response = client.text_detection(image=image)
    texts = response.text_annotations
    
    if texts:
        extracted_text = texts[0].description
    else:
        raise Exception("Metin bulunamadı.")
    
    if response.error.message:
        raise Exception(f"API Hatası: {response.error.message}")
    
    return extracted_text

def extract_info_from_text(text):
    api_key = "your gemini api key"
    client = genai.Client(api_key=api_key)
    
    response = client.models.generate_content(
        model="gemini-2.0-flash", contents=f"""
        Aşağıdaki metinden yalnızca belirli bilgileri çıkararak JSON formatında döndür. JSON’un anahtarları şu şekilde olmalı:
            - "saat": Fişte geçen saat bilgisi
            - "tarih": Fişte geçen tarih bilgisi
            - "baslik": Market ya da işletme adı
            - "adres": İşletmenin açık adresi
            - "urunler": Satın alınan ürünlerin isimleri ve fiyatları ve adetleti (JSON içinde liste olarak döndür)
            - "toplam": Satın alınan urunlerin toplam tutarı 
            Eğer herhangi bir bilgi metinde yoksa, o alanı "" (boş string) olarak döndür.
            Metin:
            
                {text}
                """
    )
    return response.text

def resize_and_show_image(image_path):
    image = cv2.imread(image_path)
    scale_percent = 50  # Boyutu %50 küçült
    width = int(image.shape[1] * scale_percent / 100)
    height = int(image.shape[0] * scale_percent / 100)
    dim = (width, height)
    
    resized = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
    cv2.imshow("Fiş Görüntüsü", resized)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def main(image_path):
    text = detect_text_from_image(image_path)
 #   print("birinci ",text)
    json_data = extract_info_from_text(text)
    print("1birinci ")
    print("Çıkarılan JSON Bilgisi:", json_data)
    resize_and_show_image(image_path)

if __name__ == "__main__":
    path=r".\realtest\v4"
    files=os.listdir(path)
    files=[file for file in files if file.endswith(".jpg")]

    main(f"{path}\\{files[2]}")
```
